**Face Detection using CV2**

```
import cv2
import matplotlib.pyplot as plt
```

Importing Image


```
img = cv2.imread('faces.jpg')
img.shape
```

Displaying image (default in BGR format)

```
plt.imshow(img)
plt.show()
```

Converting BGR color image into RGB color image

```
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

Increasing the size of the image

```
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()
```

Using haarcascades face xml file to detect the face coordinates 

```
face_xml ='face.xml'
face_cascade = cv2.CascadeClassifier(face_xml)
```

Scaling the image ( Eg: [x,y,height,weight] )

```
faces = face_cascade.detectMultiScale(img)
faces
```

Drawing the rectangle on the detected faces

Using Loop to detech all the faces on the image

```
for x,y,w,h in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
```

Displaying the detected faces

**Detecting Eyes inside the faces**

```
img = cv2.imread('faces.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_xml = 'face.xml'
eye_xml = 'eye.xml'
face_cascade = cv2.CascadeClassifier(face_xml)
eye_cascade = cv2.CascadeClassifier(eye_xml)
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
  one_face = img[y:y+h,x:x+w]   # roi - region of interest
  eyes = eye_cascade.detectMultiScale(one_face,1.3)
  for ex,ey,ew,eh in eyes:
    cv2.rectangle(one_face,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()
```

Detected the persons eyes with green square box near the eyes (Accuracy is 90-92% for the eye detection due to dark color,minimized eyes)

**Detecting smiles on faces**

```
img = cv2.imread('smiles.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_xml = 'face.xml'
smile_xml = 'smile.xml'
face_cascade = cv2.CascadeClassifier(face_xml)
smile_cascade = cv2.CascadeClassifier(smile_xml)
faces = face_cascade.detectMultiScale(img,1.2)
for x,y,w,h in faces:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
  one_face = img[y:y+h,x:x+w]   # roi - region of interest
  smiles = smile_cascade.detectMultiScale(one_face,1.4,15)
  for ex,ey,ew,eh in smiles:
    cv2.rectangle(one_face,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.show()
```

Detected the persons with green rectangle near the lips who were smiling
